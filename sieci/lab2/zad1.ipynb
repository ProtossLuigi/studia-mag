Wierzchołek o największym betweenness centrality.

```python
np.argmax(dict(nx.betweenness_centrality(G)))
```

```python
import random

def seed_graph_random(G, n=3):
    return random.sample(G.nodes, n)
def seed_betweenness(G):
    return [np.argmax(nx.betweenness_centrality(G)) + 1]
def seed_closeness(G):
    return [np.argmax(nx.closeness_centrality(G)) + 1]

def icm(G, steps=10, p=0.5, seed='3random'):
    if seed == '3random':
        active = seed_graph_random(G)
    elif seed == 'betweenness':
        active = seed_betweenness(G)
    elif seed == 'closeness':
        active = seed_closeness(G)
    else:
        raise RuntimeError()
    activated = [len(active)]

    for step in range(steps):
        new_active = []
        for edge in G.edges(active):
            if edge[1] not in active and edge[1] not in new_active and random.random() < p / G.degree(edge[1]):
                new_active.append(edge[1])
        activated.append(len(new_active))
        active += new_active
    
    return activated

def avg_icm(repeats=10, *args, **kwargs):
    results = []
    for _ in range(repeats):
        results.append(icm(*args, **kwargs))
    return np.array(results).mean(axis=0)
```

```python
results = {}
for seed in ['3random', 'betweenness', 'closeness']:
    results[seed] = {}
    for threshold in [0.1, 0.25, 0.5]:
        results[seed][threshold] = avg_icm(10, G, p=threshold, seed=seed)
```

Wykresy pokazują ilość aktywnych wierzchołków w każdym kroku Independant Cascade Model. Zostały użyte 3 metody wyboru początkowych wierzchołków: `3random` które bierze 3 losowe wierzchołki, `betweenness` bierze wierzchołek o największym betweenness i `closeness` które robi to samo tylko dla closeness.

```python
fig, ax = plt.subplots(3, 3, figsize=(24,24))
for i in range(3):
    method = list(results.keys())[i]
    for j in range(3):
        threshold = list(results[method].keys())[j]
        label = 'seed = ' + method + ', threshold = ' + str(threshold) 
        ax[i][j].plot(list(range(11)), np.cumsum(results[method][threshold]))
        ax[i][j].set_title(label)
```

Jak widać kierowanie się miarą betweenness lub closeness pozwala na szybsze rozprzestrzenianie się, nawet pomimo mniejszej ilości wierzchołków początkowych.
