```python
import pandas as pd
```

```python
df = pd.read_csv('radoslaw_email/out.radoslaw_email_email', sep='\s+', skiprows=2, header=None)[[0, 1]]
G = nx.MultiGraph()
G.add_edges_from(df.to_numpy())
```

```python
print(len(G.nodes()))
print(len(G.edges()))
```

```python
G = nx.Graph(G)
for node in G.nodes():
    if G.has_edge(node, node):
        G.remove_edge(node, node)
print(len(G.nodes()))
print(len(G.edges()))
```

```python
plt.hist(list(dict(G.degree).values()))
```

```python
plt.hist(list(nx.betweenness_centrality(G).values()))
```

```python
plt.hist(list(nx.closeness_centrality(G).values()))
```
