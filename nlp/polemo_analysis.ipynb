```python
import pandas as pd
from sklearn.metrics import cohen_kappa_score
import krippendorff
from sklearn.preprocessing import OrdinalEncoder
```

```python
zdania = pd.read_csv('polemo1_raw/zdania.tsv', sep='\t', low_memory=False)
zdania
```

```python
meta = pd.read_csv('polemo1_raw/meta.tsv', sep='\t')
meta
```

```python
annotators = ['super', 'a', 'e', 'k']
pairs = [[annotators[i], annotators[j]] for i in range(len(annotators)) for j in range(i + 1, len(annotators))]
pairs
```

```python
for pair in pairs:
    columns = zdania[pair].dropna()
    print(cohen_kappa_score(columns[pair[0]], columns[pair[1]]))
```

```python
zdania[pairs[-1]].dropna()
```

```python
for pair in pairs:
    columns = meta[pair].dropna()
    print(cohen_kappa_score(columns[pair[0]], columns[pair[1]]))
```

```python
columns = OrdinalEncoder().fit_transform(zdania[annotators])
krippendorff.alpha(reliability_data=columns.T)
```
