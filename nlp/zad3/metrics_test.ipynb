```python
import json
import pandas as pd
import numpy as np
```

```python
with open('metrics.json', 'r') as f:
    metrics = json.load(f)
with open('metrics_cnn.json', 'r') as f:
    metrics2 = json.load(f)
for k in metrics:
    metrics[k]['cnn'] = metrics2[k]['cnn']
```

```python
df = pd.DataFrame(columns=['model', 'classifier', 'precision', 'recall', 'f1'])
for v in metrics:
    for c in metrics[v]:
        row = {
            'model': v[15:-4],
            'classifier': c.upper()
        }
        best_epoch = np.argmax(metrics[v][c]['val']['f1'])
        row['precision'] = metrics[v][c]['val']['precision'][best_epoch]
        row['recall'] = metrics[v][c]['val']['recall'][best_epoch]
        row['f1'] = metrics[v][c]['val']['f1'][best_epoch]
        df = df.append(row, ignore_index=True)
```
